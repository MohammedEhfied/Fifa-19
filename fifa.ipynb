Ratio of right footed players to left footed players

```python
right = 0
left = 0
for i in df["preferred_foot"]:
    if i == "Right":
        right+=1
    else:
        left+=1
```

```python
plt.pie([right,left], labels=["Right","Left"],explode =(0.1,0),shadow=True,autopct='%.2f%%',startangle=50)
plt.show()
```

Goalkeepers' ratio of right footed players to left footed players

```python
right_gk = 0
left_gk = 0
for i in df["id"]:
    if df["preferred_foot"][i] == "Right" and df["positions"][i] == "GK":
        right_gk+=1
    elif df["preferred_foot"][i] == "Left" and df["positions"][i] == "GK":
        left_gk+=1
```

```python
plt.pie([right_gk,left_gk], labels=["Right","Left"],explode =(0.1,0.3),shadow=True,autopct='%.2f%%',startangle=50)
plt.show()
```
